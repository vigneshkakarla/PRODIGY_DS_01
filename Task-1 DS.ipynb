```python
import matplotlib.pyplot as plt

# Sample data (replace with real population data)
years = [2000, 2010, 2020, 2030]
population = [1053000000, 1210000000, 1380000000, 1500000000]

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population (in billions)')
plt.title('Population in India Over the Years')
plt.xticks(years)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the bar chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data (replace with your own dataset)
data = np.random.normal(0, 1, 1000)  # Generates 1000 random values with a mean of 0 and standard deviation of 1

# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the histogram
plt.show()
```
